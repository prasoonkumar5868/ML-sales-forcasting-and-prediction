```python
df['Order Date'] = pd.to_datetime(df['Order Date'])  # Time based features
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
df['Week'] = df['Order Date'].dt.isocalendar().week
df['Day'] = df['Order Date'].dt.day
df['DayOfWeek'] = df['Order Date'].dt.dayofweek  # 0 = Monday
df['IsWeekend'] = df['DayOfWeek'].isin([5, 6])
```

```python
df['Product Category'] = df['Product Category'].str.strip().str.title() #standarising names
df['Payment Type'] = df['Payment Type'].str.strip().str.title()
df['Order Status'] = df['Order Status'].str.strip().str.title()
```

```python
df = df[(df['Quantity'] > 0) & (df['Quantity'] < df['Quantity'].quantile(0.99))]  #removing unusually large or small values
df = df[(df['Unit Price'] < df['Unit Price'].quantile(0.99))]
```

```python
df.shape
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Group total sales by day
daily_sales = df.groupby('Order Date')['Total Price'].sum().reset_index()

# Plot
plt.figure(figsize=(14, 6))
sns.lineplot(data=daily_sales, x='Order Date', y='Total Price')
plt.title('Daily Sales Trend')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Extract month and year
df['Order Date'] = pd.to_datetime(df['Order Date'])  
df['YearMonth'] = df['Order Date'].dt.to_period('M').astype(str)

# Group total sales by Year-Month
monthly_sales = df.groupby('YearMonth')['Total Price'].sum().reset_index()

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(data=monthly_sales,
    x='YearMonth',
    y='Total Price',
    hue='YearMonth',             
    palette='Blues_d',
    legend=False  )
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Add DayOfWeek and Weekend flag
df['DayOfWeek'] = df['Order Date'].dt.dayofweek  # Monday=0, Sunday=6
df['IsWeekend'] = df['DayOfWeek'].isin([5, 6])    # Saturday or Sunday

# Group by weekend flag
weekend_sales = df.groupby('IsWeekend')['Total Price'].sum().reset_index()

# Rename for clarity
weekend_sales['Day Type'] = weekend_sales['IsWeekend'].map({True: 'Weekend', False: 'Weekday'})

# Plot
plt.figure(figsize=(8, 5))
sns.barplot(data=weekend_sales, x='Day Type', y='Total Price')
plt.title('Total Sales: Weekday vs Weekend')
plt.xlabel('')
plt.ylabel('Total Sales')
plt.tight_layout()
plt.show()
```

```python
# Group total sales by product category
category_sales = df.groupby('Product Category')['Total Price'].sum().reset_index().sort_values(by='Total Price', ascending=False)

# Plot
plt.figure(figsize=(12, 6))
sns.lineplot(data=category_sales, x='Product Category', y='Total Price')
plt.title('Total Sales by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
payment_counts = df['Payment Type'].value_counts().reset_index()
payment_counts.columns = ['Payment Type', 'Order Count']

# Sort by count for nicer line shape
payment_counts = payment_counts.sort_values('Order Count')

# Line plot
plt.figure(figsize=(10, 5))
sns.lineplot(data=payment_counts, x='Payment Type', y='Order Count', marker='o')
plt.title('Payment Method Usage (Line Plot)')
plt.xlabel('Payment Type')
plt.ylabel('Number of Orders')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Unit Price', bins=50, kde=True, color='skyblue')
plt.title('Distribution of Unit Prices')
plt.xlabel('Unit Price')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
numeric_df = df.select_dtypes(include='number')

# Compute correlation matrix
corr_matrix = numeric_df.corr(numeric_only=True)

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Numeric Features')
plt.tight_layout()
plt.show()
```
