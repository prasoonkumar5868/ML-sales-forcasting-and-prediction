```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from prophet import Prophet
import pycountry_convert as pc
from sklearn.metrics import mean_absolute_percentage_error
```

```python
df = pd.read_csv("Agrofood_co2_emission.csv")
```

FORCASTING


```python
# Map countries to continents
def get_continent(country_name):
    try:
        country_code = pc.country_name_to_country_alpha2(country_name)
        continent_code = pc.country_alpha2_to_continent_code(country_code)
        return pc.convert_continent_code_to_continent_name(continent_code)
    except:
        return 'Unknown'

df['Continent'] = df['Area'].apply(get_continent)
df = df[df['Continent'] != 'Unknown']

# Aggregate by continent and year
continent_emissions = df.groupby(['Continent', 'Year'])['total_emission'].sum().reset_index()

# Containers for results
forecast_df = pd.DataFrame()
future_forecast_df = pd.DataFrame()
mape_scores = {}

# Forecast + Evaluate
for continent in continent_emissions['Continent'].unique():
    data = continent_emissions[continent_emissions['Continent'] == continent]
    data = data.rename(columns={'Year': 'ds', 'total_emission': 'y'})
    data['ds'] = pd.to_datetime(data['ds'], format='%Y')
    data = data.sort_values('ds')

    # Train-test split: last 5 years as test
    train = data.iloc[:-5]
    test = data.iloc[-5:]

    # Prophet model training on train set
    model = Prophet(yearly_seasonality=True)
    model.fit(train)

    # Predict including test years
    future_test = model.make_future_dataframe(periods=5, freq='Y')
    forecast_test = model.predict(future_test)
    forecast_test['Continent'] = continent

    # Evaluate on test
    forecast_test['year'] = forecast_test['ds'].dt.year
    test['year'] = test['ds'].dt.year
    merged = forecast_test[['year', 'yhat']].merge(test[['year', 'y']], on='year')
    mape = mean_absolute_percentage_error(merged['y'], merged['yhat']) * 100
    mape_scores[continent] = round(mape, 2)

    # Save prediction
    forecast_df = pd.concat([forecast_df, forecast_test[['ds', 'yhat', 'Continent']]])

    # Train on full data for final 5-year forecast
    model_final = Prophet(yearly_seasonality=True)
    model_final.fit(data)
    future = model_final.make_future_dataframe(periods=5, freq='Y')
    future_forecast = model_final.predict(future)
    future_forecast['Continent'] = continent
    future_forecast_df = pd.concat([future_forecast_df, future_forecast[['ds', 'yhat', 'Continent']]])




```

```python
# Plot future forecast (raw)
future_forecast_df['Year'] = future_forecast_df['ds'].dt.year
last_year = continent_emissions['Year'].max()
plot_df = future_forecast_df[future_forecast_df['Year'] > last_year]

plt.figure(figsize=(12, 6))
for continent in plot_df['Continent'].unique():
    sub_df = plot_df[plot_df['Continent'] == continent]
    plt.plot(sub_df['Year'], sub_df['yhat'], label=continent)

plt.title("Forecasted CO₂ Emissions by Continent (Unit Unknown, Predicted 2021–2025)")
plt.xlabel("Year")
plt.ylabel("Predicted Emissions")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()


# Plot normalized forecast (to show trend)
norm_plot_df = plot_df.copy()
norm_plot_df['yhat_norm'] = norm_plot_df.groupby('Continent')['yhat'].transform(lambda x: x / x.max())

plt.figure(figsize=(12, 6))
for continent in norm_plot_df['Continent'].unique():
    sub_df = norm_plot_df[norm_plot_df['Continent'] == continent]
    plt.plot(sub_df['Year'], sub_df['yhat_norm'], label=continent)

plt.title("Normalized Forecasted CO₂ Emissions (Relative Trends)")
plt.xlabel("Year")
plt.ylabel("Normalized Emissions (0-1)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Print evaluation
print("MAPE scores (test set, last 5 years):")
for k, v in mape_scores.items():
    print(f"{k}: {v}%")
```
